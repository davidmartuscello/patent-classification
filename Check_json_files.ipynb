```python
import pandas as pd
import glob
```

```python
jsonfiles = [f for f in glob.glob("..\json_files_3\*.json")]
```

```python
temp = pd.DataFrame()
df1 =  pd.DataFrame()
```

```python
for f in jsonfiles:
    temp = pd.read_json(path_or_buf=f, orient = 'values')
    df1 = pd.concat([df1, temp], ignore_index=True)
```
